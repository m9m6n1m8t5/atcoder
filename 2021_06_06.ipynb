```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------
import numpy as np

N = int(input())
T = list(map(int, input().split()))
S = sum(T)
half = S//2+1
dp = [[0]*half for _ in range(N+1)]
dp[0][0] = 1
for i in range(N):
    for j in range(half):
        dp[i+1][j] |= dp[i][j]
        if j+T[i] < half:
            dp[i+1][j+T[i]] |= dp[i][j]
for i in reversed(range(half)):
    if dp[-1][i]:
        print(S-i)
        break
```
