## D

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------

H, W = map(int, input().split())
A_tmp = [list(input()) for _ in range(H)]
A = [[0]*W for _ in range(H)]
for h in range(H):
    for w in range(W):
        if (h+w)%2==0:
            A[h][w] = 1 if A_tmp[h][w]=="-" else -1
        else:
            A[h][w] = 1 if A_tmp[h][w]=="+" else -1
A[0][0] = 0

table = [[0]*W for _ in range(H)]

table[-1][-1] = A[-1][-1]

for w in reversed(range(W-1)):
    table[-1][w] = table[-1][w+1] + A[-1][w]

for h in reversed(range(H-1)):
    table[h][-1] = table[h+1][-1] + A[h][-1]
    
for h in reversed(range(H-1)):
    for w in reversed(range(W-1)):
        if (h+w)%2 == 0:
            table[h][w] = max(table[h+1][w], table[h][w+1]) + A[h][w]
        else:
            table[h][w] = min(table[h+1][w], table[h][w+1]) + A[h][w]

if table[0][0]>0:
    print("Takahashi")
elif table[0][0]<0:
    print("Aoki")
else:
    print("Draw")
table
```
