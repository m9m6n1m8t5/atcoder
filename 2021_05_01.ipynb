```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------

N = int(input())
M = [tuple(map(int, input().split())) for _ in range(N)]

def judge(x):
    s = set()
    for m in M:
        s.add(sum(1<<i for i in range(5) if m[i]>=x))
    for x in s:
        for y in s:
            for z in s:
                if x|y|z == 31:
                    return True
    return False

left = 0
right = 10**9 + 1
count = 0
while right - left > 1:
    count += 1
    mid = (right + left) //2
    if judge(mid):
        left = mid
    else:
        right = mid
print(left)
```

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------
from collections import deque

S = list(input())
T = deque()

flag = False
for c in S:
    if c == 'R':
        flag = not flag
    else:
        if flag:
            if T and c==T[0]:
                T.popleft()
            else:
                T.appendleft(c)
        else:
            if T and c==T[-1]:
                T.pop()
            else:
                T.append(c)
if flag:
    T = reversed(T)

T = "".join(T)
print(T)
```
