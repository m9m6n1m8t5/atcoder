## 001	Yokan Party（★4）『2分探索』

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------

N, L = map(int, input().split())
K = int(input())
A = list(map(int, input().split()))

def solve(x):
    pre = 0
    count = 0
    for i in range(N):
        if  A[i] - pre>= x and L - A[i] >= x:
            count += 1
            pre = A[i]
    if count >= K:
        return True
    return False

left, right = 0, L
while right - left > 1:
    mid = (left + right) // 2
    if solve(mid):
        left = mid
    else:
        right = mid
print(left)
```

## 002	Encyclopedia of Parentheses（★3）『bit全探索』

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------

def judge(s):
    count = 0
    for c in s:
        count += 1 if c=='(' else -1
        if count < 0:
            return False
    return count==0

def main():
    N = int(input())
    
    ans = []
    for n in range(2**N):
        brackets = []
        for i in reversed(range(N)):
            if (n>>i) & 1 == 0:
                brackets.append("(")
            else:
                brackets.append(")")
        brackets =  "".join(brackets)
        if judge(brackets): ans.append(brackets)
    print("\n".join(ans))

main()
```

## 003	Longest Circular Road（★4）『木の直径』

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------
import time
from collections import deque

def BFS(node, graph):
    q = deque()
    q.append(node)
    dist = [-1] * len(graph)
    dist[node] = 0
    
    while q:
#         print(q)
        current_node = q.popleft()
        for i in graph[current_node]:
            if dist[i] == -1:
                q.append(i)
                dist[i] = dist[current_node] + 1
    return dist

def argmax(array):
    max_i = 0
    max_v = array[0]
    for i, v in enumerate(array):
        if v > max_v:
            max_v = v
            max_i = i
    return max_i
def main():
    N = int(input())
#     graph = [[0] * N for _ in range(N)]
    graph = {i: set() for i in range(N)}
    for i in range(N-1):
        A, B = map(int, input().split())
        graph[A-1].add(B-1)
        graph[B-1].add(A-1)
#         graph[A-1][B-1] = 1
#         graph[B-1][A-1] = 1
    deepest = argmax(BFS(0, graph))
    print(max(BFS(deepest, graph)) + 1)
start = time.time()    
main()
print(time.time() - start)
```

## 004	Cross Sum（★2）『』

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------

H, W = map(int, input().split())
A = [list(map(int, input().split()))for _ in range(H)]
sum_row = [0] * W
sum_col = [0] * H
for i in range(H):
    for j in range(W):
        sum_row[j] += A[i][j]
        sum_col[i] +=A[i][j] 

for i in range(H):
    for j in range(W):
        A[i][j] = sum_row[j] + sum_col[i] - A[i][j]
for row in A:
    print(*row)
```

## 006 - Smallest Subsequence（★5）『貪欲法』

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------

N, K = map(int, input().split())
S = input()
matrix = [[N] * 26 for _ in range(N)]
matrix[N-1][ord(S[-1])-ord('a')] = N-1
for i in reversed(range(N-1)):
    c = ord(S[i]) - ord('a')
    for j in range(26):
        if c == j:
            matrix[i][j] = i
        else:
            matrix[i][j] = matrix[i+1][j]
i = 0
ans = []
while i < N:
    for j in range(26):
        if K-1 < N - matrix[i][j]:
            ans.append(chr(ord('a') + j))
            K -= 1
            i = matrix[i][j] + 1
            break
    if K <= 0:
        break
print("".join(ans))
```

## 007 - CP Classes（★3）『ソートして二分探索』

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------

N = int(input())
A = sorted(list(map(int, input().split())))
Q = int(input())
B = [int(input()) for _ in range(Q)]

score = []
for b in B:
    left = 0
    right = N
    while right - left > 1:
        mid = (left + right)//2
        if A[mid] <= b:
            left = mid
        else:
            right = mid
    diff1 = abs(A[left]-b)
    diff2 = abs(A[left+1]-b)if left<N-1 else diff1
    score.append(min(diff1, diff2))
print("\n".join(map(str, score)))
```

## 008 - AtCounter（★4）

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------

N = int(input())
S = input()
mod = 10**9+7
atcoder = "atcoder"
table = [[1]*(N+1)]
table.extend([[0]*(N+1) for _ in range(len(atcoder))])

for i in range(1, len(atcoder)+1):
    for j in range(1, N+1):
        table[i][j] = (table[i][j-1] + (table[i-1][j-1] if S[j-1]==atcoder[i-1] else 0)) % mod

print(table[-1][-1])
```
