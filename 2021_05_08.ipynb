## D

```python
IN = iter(Input.split("\n")).__next__
def input():
    return IN()

# ------------------------------------------------------
from itertools import combinations

N = int(input())
A = list(map(int, input().split()))

table = [[] for _ in range(200)]
flag = False
for i in range(1, N):
    if not flag:
        for c in combinations(range(N), i):
            mod = sum([A[k] for k in c])%200
            table[mod].append(c)
            if len(table[mod]) >= 2:
                flag = True
                break

if flag:
    print("YES")
    x = [x+1 for x in table[mod][0]]
    print(len(x), " ".join(map(str,x)))
    y = [y+1 for y in table[mod][1]]
    print(len(y), " ".join(map(str,y)))
else:
    print("NO")
```

```python
from ipywidgets import Textarea
def get_input(change):
    global Input
    Input = change["new"]

textarea = Textarea()
textarea.observe(get_input, names="value")
display(textarea)
```
